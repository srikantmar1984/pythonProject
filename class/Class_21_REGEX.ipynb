```python
# Personal errror
class My_exception(Exception):
    def __init__(self , msg):
        self.msg = msg
```

```python
def amount_deposited(amt):
    if amt > 50000 and amt < 60000:
        raise My_exception('Overlimit')
    elif amt > 60000:
        print('Cant transfer this amount')
    else:
        print('amount deposited')
```

```python
def submit_pan(pan):
    if len(pan) == 10:
        print('we will verify soon')
    else:
        raise My_exception("Please enter correct pan number ")
```

```python
amt = int(input('Please enter the amount you need to transfer : '))

try:
    amount_deposited(amt)
except My_exception as e:
    print(e)
    try:
        pan = input('Enter the pan number ')
        submit_pan(pan)
    except My_exception as e:
        print(e)
```

# REGEX

* if we want to find the particular or specific things from the string we use use

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```
