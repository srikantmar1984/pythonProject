```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
df
```

```python
for i in df.index:
    print(i)
```

```python
for i in df.index:
    if df.loc[i , 'Organizations_name'] == 'ISRO':
        print(df.loc[i , 'Headquarters'])
```

```python
for row in df.itertuples():
    if row.Organizations_name == 'ISRO':
        print(row.Country_name)
```

```python
# Iter rows


for row in df.iterrows():
    print(row)
```

```python
# Iter rows


for row in df.iterrows():
    print(row[1] , end = '\n\n\n')
```

# Concatination

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
df = pd.concat([india_weather , us_weather])
df
```

```python
# 1 way
df = pd.concat([india_weather , us_weather] , ignore_index= True)
df
```

```python
# 2nd way

df.reset_index()
```

```python
# 2nd way

df.reset_index(drop= True ,inplace = True)
df
```

# Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

# Merge Continue

```python
df_1 = pd.concat([df1 , df2])
df_1
```

```python
df_2 = pd.merge(df1 ,df2 , how = 'outer')
df_2
```
