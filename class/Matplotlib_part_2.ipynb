```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,45,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp
totallist=[max_t,min_t,avg_t]

for t in totallist:
    plt.plot(days,t)
    plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature vs Day')
plt.legend(['max', 'min','avg'], loc = 'upper right', ncol=3)
```

```python
blood_sugar = [113 , 85 ,90,150 , 149, 89,93,100,89,98,91,129]


# 1. How may people are normal
# 2. How may people are prediabetic
# 3. How may people have diabetics

# 80-100 - normal
# 100-125 - prediabetic
# 125 - diabetics
```

```python
# 

plt.hist(blood_sugar) #bins = 10
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]


plt.hist([blood_sugar_men , blood_sugar_women] ,
         bins = [67 ,100 , 125, 150] ,color = ['g' , 'orange'],
        label = ['Men' , 'Women'])
plt.legend()
```

# Subplots

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400]

plt.subplot(2,2,1)
plt.bar(year , income)
plt.title('hello')

plt.subplot(2,2,4)
plt.bar(year , expense)
plt.title('xyz')
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400]

plt.subplot(2,2,1)
plt.bar(year , income)
plt.title('hello')

plt.subplot(2,2,2)
plt.pie(income , labels = year)

plt.subplot(2,2,4)
plt.bar(year , expense)
plt.title('xyz')
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400]

plt.subplot(2,2,1)
plt.bar(year , income)
plt.title('hello')

plt.subplot(2,2,2)
plt.pie(income , labels = year)

plt.subplot(2,2,4)
plt.bar(year , expense)
plt.title('xyz')
plt.savefig('SUBPLOT_EXAMPLE')
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400]

plt.subplot(2,2,1)
plt.bar(year , income ,animated = True)
plt.title('hello')

plt.subplot(2,2,2)
plt.pie(income , labels = year)

plt.subplot(2,2,4)
plt.bar(year , expense)
plt.title('xyz')
plt.savefig('SUBPLOT_EXAMPLE.pdf')
```
