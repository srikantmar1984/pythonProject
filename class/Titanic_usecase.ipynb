```python
# z-score
mean_of_fare = np.mean(df['fare'])
std_of_sales = np.std(df['fare'])

threshold = 3
outlier = []


for i in df['fare']:
    z = (i-mean_of_fare) / (std_of_sales)
    if z > threshold:
        outlier.append(i)

        
print('Total no of outliers in your dataset are : ' , len(outlier))
print('Maximum fare outlier : ' , max(outlier))
print('Minimum fare outlier : ' , min(outlier))
```

```python
sns.displot(x = outlier , bins = 20).set(title = 'Outlier distribution of fare')
```

```python
to_drop = df.loc[df['fare'] > min(outlier)]
print(len(to_drop))
print(to_drop)
```
