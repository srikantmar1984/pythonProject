syntax

# iloc 

- We can do slicing based rows ans columns
- We can do it by index values 


syntax
datafram.iloc[row_name_start : row_stop : step ,column_start : columns_end : step]

- Zero indexing is not possible with loc

```python
new_bike_data
```

```python
new_bike_data.loc[1:4 , 0:4]
```

```python
bike_share.loc[0:5 ,0:4]
```

```python
new_bike_data.iloc[1:4 , 0:4]
```

```python
bike_share.iloc[0:5 ,0:4]
```

```python
# normal slicing
- index and custom row slicing is possible
- Cannot access columns


# loc
- Both rows and columns are accessable
- you have to call the rows and columns by names


# iloc
- Both rows and columns are accessable
- you can call both rows and columns by index ad well as name

```
